```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
job = pd.read_csv('/home/sarthak/Downloads/combinejob.csv')
```

```python
job
```

```python
job.drop(['missing info', 'job name' , 'logical job name'], axis=1)
```

```python
job.rename(columns={'event type':'event_type','user name':'user_name','job ID':'job_id'}, inplace=True)
```

```python
job
```

```python
job.timestamp= np.where( job.timestamp > 800, job.timestamp/(1000000), job.timestamp)
job
```

```python
pd.to_datetime(	2.501193e+06, unit='s',origin=pd.Timestamp('2011-01-01'))
```

```python
job['date'] = pd.to_datetime(job['timestamp'],unit='s',origin=pd.Timestamp('2011-01-01'))
```

```python
job
```

```python
job['just_date'] = job['date'].dt.date
```

```python
job
```

```python
job=job.sort_values(by=['just_date'])
```

```python
job_failed=job.loc[(job['event_type'] == 3)|(job['event_type'] == 5)|(job['event_type'] == 2)]
```

```python
job_failed
```

```python
job_failed['day']= job_failed['date'].dt.day
job_failed
```

```python
failed=job_failed.groupby(["day"]).size().reset_index(name="freq")
failed
```

```python
sns.regplot(x="day",y="freq",data=failed)
```

```python
failed = failed.set_index('day')
```

```python
plt.style.use('fivethirtyeight')
failed.plot()
plt.show()
```
