```python
event_job=job_1.groupby(["event type"]).size().reset_index(name="total")
```

```python
event_job
```

```python
event_job=event_job.drop(event_job.index[[0,1]])
event_job
```

```python
usage = event_job['total']
labels = event_job['event type']
# Generating the y positions. Later, we'll use them to replace them with labels.
y_positions = range(len(labels))

# Creating our bar plot
plt.bar(y_positions, usage)
plt.xticks(y_positions, labels)
plt.ylabel("total jobs")
plt.title("jobs per event type")
plt.show()
```

```python
Total = event_job['total'].sum()
Total
```

```python
event_job['perc'] = 100*event_job['total']/Total
event_job
```

```python
Y = event_job['perc']
X = event_job['event type']
# Generating the y positions. Later, we'll use them to replace them with labels.
y_positions = range(len(labels))

# Creating our bar plot
plt.bar(y_positions, Y)
plt.xticks(y_positions, X)
plt.ylabel("total jobs percentage")
plt.title("jobs per event type")
plt.show()
```
